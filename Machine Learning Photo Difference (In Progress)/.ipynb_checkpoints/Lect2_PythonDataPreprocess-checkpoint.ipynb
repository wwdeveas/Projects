## Eliminate outliers using box plot

Use numpy to get the q1 and q3 quartile

```python
print (data.Age.quantile(q=[0.25, 0.75]))
```

```python
import numpy as np

# get q1 and q3 quartile in numbpy
q1 = np.quantile (data['Age'], 0.25)
q3 = np.quantile (data['Age'], 0.75)
print (q1, q3)
iqr = q3 - q1
 
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr

outlier = []
for d in data['Age']:
    if d < lower_limit or d > upper_limit: outlier.append (d)
print (outlier)
if outlier != []:
    for d in outlier:
        df2 = data [data['Age'] != d]
```

## Standard deviation method to detect and remove outlier

```python
# Function to Detection Outlier on one-dimentional datasets.
def find_anomalies(dataColumn):
    #define a list to accumlate anomalies
    outliers = []
    
    # Set upper and lower limit to 3 standard deviation
    data_std = dataColumn.std()
    data_mean = dataColumn.mean()
    outlier_cut_off = data_std * 3
    
    lower_limit  = data_mean - outlier_cut_off 
    upper_limit = data_mean + outlier_cut_off

    # Generate outliers
    for outlier in dataColumn:
        if outlier > upper_limit or outlier < lower_limit:
            outliers.append(outlier)
    return outliers

outliers = find_anomalies(data['Age'])
print('outliers=', outliers)
if outliers != []:
    for d in outliers:
        df2 = data [data['Age'] != d]
```

## Handling categorical data

Two types:
* Nomial categorical data: that is no intrinsic ordering between categorical data, e.g. red, blue, and green
* Ordinal categorical data: that is natural ordering between categorical data, e.g. low, medium, high

### 1. Nominal Cateorical data:

```python
import pandas as pd
data = pd.read_csv('titanic.csv')
data.dropna (subset = ['Embarked', 'Age'], axis=0, inplace=True)  # Remove missing values
data.drop (columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True) # Select features
data.head()
```

```python
from sklearn.preprocessing import LabelBinarizer
# Create one-hot encoder
one_hot = LabelBinarizer()
# One-hot encode feature
a = pd.DataFrame (one_hot.fit_transform(data['Embarked']), columns=one_hot.classes_)
df2 = data.join (a)

df2.drop (columns=['Embarked'], axis=0, inplace=True)
df2.head()
```

```python
a = pd.get_dummies (data, columns=['Embarked', 'Sex'])
a.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

#data['Embarked']=data['Embarked'].astype('category')

enc=OneHotEncoder()
a=enc.fit_transform(data[['Embarked']]).toarray()
enc_data=pd.DataFrame(a, columns=list(enc.categories_[0]))
df2 = data.join (enc_data)
df2.drop (columns=['Embarked'], axis=0, inplace=True)
df2.head()
```

## Imbalanced data

```python
# Load libraries
import numpy as np
from sklearn.datasets import load_iris

# Load iris data
iris = load_iris()

# Create feature matrix
features = iris.data

# Create target vector
target = iris.target

# Remove first 40 observations
features = features[40:,:]
target = target[40:]

# Create binary target vector indicating if class 0
target = np.where((target == 0), 0, 1)

# Look at the imbalanced target vector
target
```

```python
from sklearn.ensemble import RandomForestClassifier
# Create weights
weights = {0: .9, 1: 0.1}

# Create random forest classifier with weights
RandomForestClassifier(class_weight=weights)
```

### Down-Sample:we randomly sample without replacement from the majority class to create a new subset of observations equal in size to the minority class.

```python
# Indicies of each class' observations
i_class0 = np.where(target == 0)[0]
i_class1 = np.where(target == 1)[0]

# Number of observations in each class
n_class0 = len(i_class0)
n_class1 = len(i_class1)

# For every observation of class 0, randomly sample
# from class 1 without replacement
i_class1_downsampled = np.random.choice(i_class1, size=n_class0, replace=False)

np.concatenate((target[i_class0], target[i_class1_downsampled]))
```

### Up-Sample:we randomly sample without replacement from the minority class to create a new subset of observations equal in size to the majority class.

```python
# Indicies of each class' observations
i_class0 = np.where(target == 0)[0]
i_class1 = np.where(target == 1)[0]

# Number of observations in each class
n_class0 = len(i_class0)
n_class1 = len(i_class1)

# For every observation of class 0, randomly sample
# from class 1 without replacement
i_class0_upsampled = np.random.choice(i_class0, size=n_class1, replace=True)

np.concatenate((target[i_class0_upsampled], target[i_class1]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = df2.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
#Correlation with output variable
cor_target = abs(cor["medv"])
threshold = 0.3
#Selecting highly correlated features
relevant_features = cor_target[cor_target>threshold]
relevant_features
```

```python
# Keep relevant features and remove the rest:
keepFeatures = []
for a in relevant_features.keys():
    keepFeatures.append (a)
df = df[keepFeatures]
df.head()
```

```python
for f1 in df.columns:
    if f1 == 'Unnamed: 0': continue
    for f2 in df.columns:
        if f2 == 'Unnamed: 0': continue
        if f1 == f2: continue
        c = df[[f1, f2]].corr()
        if c.loc[f1, f2] > 0.9: print (f1, f2, c.loc[f1, f2])
```
